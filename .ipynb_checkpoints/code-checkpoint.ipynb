```python
phrase = "toto sale Suisse"
mots = phrase.split(" ")
```

```python
print(mots)
```

```python
eids = []

for mot in mots: 
    eids.append(int(E.loc[E['name'] == mot]['eid']))
    
print(eids)
```

```python
for eid in eids: 
    print(R.loc[R['node1'] == str(eid)].loc[R['node2'] == str(eid)]['w'])
```
